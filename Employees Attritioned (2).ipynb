### Finding highest and lowest salries

```python
Highest_salary=data['MonthlyIncome'].max()
print(Highest_salary)
lowest_salary=data['MonthlyIncome'].min()
print(lowest_salary)
```

### Grouping Attrition and Monthly income columns

```python
grouped_data=data.groupby('Attrition')['MonthlyIncome'].mean().reset_index()
print(grouped_data)
```

### Visualization of attrition vs Monthly income

```python
plt.bar(grouped_data['Attrition'],grouped_data['MonthlyIncome'])
plt.title("Monthly Income VS Atttrition")
plt.ylabel("Average MonthlyIncome")
plt.xlabel("Attrition")
plt.grid(True)
plt.show()
```

### Grouped Department and Attrition columns

```python
grouped_data=data.groupby(['Department','Attrition']).size().reset_index(name='Count')
print(grouped_data)
```

### Summarized the Department and Attrtion Columns using pivot 

```python
pivot_data=grouped_data.pivot(index='Department',columns='Attrition',values='Count')
print(pivot_data)
```

### Visualization of Department and Attrition

```python
pivot_data.plot(kind='bar')
plt.title('Attrition by Department')
plt.ylabel('Count of employees')
plt.xlabel('Department')
plt.legend(title='Attrition')
plt.show()
```
